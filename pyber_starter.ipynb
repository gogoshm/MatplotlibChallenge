```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data= "data/city_data.csv"
ride_data = "data/ride_data.csv"

# Read the City and Ride Data
city = pd.read_csv(city_data)
ride = pd.read_csv(ride_data)

# Combine the data into a single dataset
data = pd.merge(city, ride, on="city")
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
# Build the scatter plots for each city types
# Incorporate the other graph properties
# Create a legend
# Incorporate a text label regarding circle size
# Save Figure
```

```python
#Average Fare ($) Per City
#Total Number of Rides Per City
#Total Number of Drivers Per City
#City Type (Urban, Suburban, Rural)

group = data.groupby(["type", "city"])

avg_fare = round(group["fare"].mean(),2)
total_ride = group["ride_id"].count()
total_driver = group["driver_count"].sum()
```

```python
plt.grid()
plt.title("Pyber Ride Sharing Data(2016)")
plt.xlabel("Total Number of Rides(Per City)")
plt.ylabel("Average Fare ($)")
plt.scatter(total_ride,avg_fare,s=total_driver,alpha=0.75, c=["red","blue", "green"])

plt.show()
plt.xlim(5,45)
plt.ylim(15,45)
plt.tight_layout()
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
# Build Pie Chart
# Save Figure
group_fare = data.groupby(["type"])
total_fare = group_fare["fare"].sum()
total_fare
```

```python
# Show Figure
fare_color = ["red", "blue", "green"]
fare_type = ["Rural", "Suburban", "Urban"]
fare_explode =(0,0.05, 0.01)
plt.title("% of Total Fares by City Type")
plt.pie(total_fare, explode=fare_explode, labels=fare_type, colors=fare_color, autopct="%1.1f%%", startangle=90)
plt.axis("equal")
plt.tight_layout()
plt.show()
```
